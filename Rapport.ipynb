Introduktion

I den här uppgiften har jag arbetat med multipel linjär regression för att analysera bostadspriser i Kalifornien år 1997. Syftet har varit att bygga modellen själv från grunden och samtidigt göra en fullständig statistisk analys av den.


Metod

Jag började med att läsa in datan och undersöka vilka variabler som fanns. Därefter valde jag vilka features som skulle ingå i modellen. För att undvika problem med att variabler är för lika varandra beräknade jag Pearsons korrelationstal mellan dem.

Själva regressionsmodellen byggdes enligt formeln för Ordinary Least Squares (OLS). Koefficienterna beräknades med hjälp av en pseudoinvers, vilket ger än stabilare modell även om vissa variabler är starkt korrelerade.

Efter att modellen tränats beräknade jag:
Antal observationer (n)
Antal variabler (d)
Residualer
Varians
RMSE
R²

För att undersöka om modellen är statistiskt säker använde jag:
F-test för hela modellen
t-test för varje enskild variabel
Konfidensintervall för parametrarna

Jag testade även olika kombinationer av variabler och valde den modell som gav högst förklaringsgrad utan att för den delen ha för hög intern korrelation.


Resultat

Resultatet visar att modellen är statistiskt signifikant. F-testet gav ett lågt p-värde, vilket betyder att modellen som helhet har förklaringsvärde.

R²-värdet visar att modellen kan förklara en stor del av den variation som finns i huspriserna. RMSE ger en uppskattning av hur stora felen i genomsnitt är.

Vid analys av enskilda variabler visade t-testen att vissa variabler, särskilt median income, har ett starkt samband med huspriset.

Konfidensintervallen visar att skattningarna är relativt stabila.


Diskussion

Den här modellen beskriver sambandet mellan olika faktorer och huspriser för ett specifikt år. Den är statistiskt fungerande och uppfyller kraven som fanns i uppgiften.

En tanke som slog mig var om det på något vis gick att använda modellen för att få fram prisutveckling över tid men det var svårt då granddatan inte hade något tidsperspektiv att använda sig av utan alla observationer var från 1997. Tillgänglig data visar bara på strukturella samband och inte framtida prognoder.
